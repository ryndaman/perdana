```python
import openpyxl
from reportlab.lib.pagesizes import letter
from reportlab.platypus import SimpleDocTemplate, Image
from reportlab.platypus import Spacer

# Load the Excel workbook
workbook = openpyxl.load_workbook('SpacePrimaryState.xlsx')

# Access a specific worksheet
worksheet = workbook['Sheet1']

# Initialize the PDF document
pdf_filename = 'output.pdf'
doc = SimpleDocTemplate(pdf_filename, pagesize=letter)

# Initialize a list to store PDF content
pdf_content = []

# Get the last 8 rows of the worksheet
rows = list(worksheet.iter_rows(min_row=2))[-8:]

# Define the path to the directory where your images are stored
image_directory = 'C:\\Users\\User\\OneDrive - PT. Nadava Sukma Lestari\\CODES\\excel script\\LAPORAN HARIAN\\FOTOVIDEO PEKERJAAN 2'

# Iterate through the selected rows
for row in rows:
    cell_data = [cell.value for cell in row]
    local_path = cell_data[10]  # Assuming local paths are in the 12th column (0-based index)

    # Combine the image directory path with the local path
    image_path = os.path.join(image_directory, local_path)

    try:
        # Add the image to the PDF
        img = Image(image_path, width=200, height=200)
        pdf_content.append(img)

        # Add some space between images
        pdf_content.append(Spacer(1, 12))
    except Exception as e:
        print(f"Error processing image from {image_path}: {str(e)}")

# Build the PDF document
doc.build(pdf_content)
```
